### モデルの学習状況(学習不足・過学習)を見る
#### Learning Curve
x軸に学習回数、y軸に正解率をとる。

上の線が訓練データ、下がテストデータ。

正しく学習が行われていれば、学習回数が増えるに連れて、訓練データのグラフは右肩下がり、テストデータのグラフは右肩上がりに変化する。（訓練データとテストデータの正解率が近づいていく）

#### *学習回数が増えても訓練データ・テストデータ両方に対する正解率が上がらない場合*


学習不足を起こしている。説明変数の特徴量を増やすことで改善することができる。



#### *訓練データの正解率が高いまま、テストデータに対する正解率が上がらない場合*



訓練データに対する過学習を起こしている。データ数が少なく、特徴量が多い場合に過学習を起こすことが多いので、特徴量を減らすか、データ数を増やすことで解決する。




````
plot_model(rf, 'learning')
````

### Validation Curve(検証曲線)
検証曲線は、ハイパーパラメーターの値と予測性能の関係を示した折れ線グラフである。x軸は、ハイパーパラメーターの値であり、縦軸は評価指標である。評価指標は、訓練データを使った時の評価指標と検証データを使った時の評価指標の両方をプロットする。両者の指標で描かれた曲線の離れ具合で、最適なハイパーパラメーターの取りうる値を大まかに把握することができる。

例）グラフが右肩下がり
  
  パラメーターの値が大きくなるにつれて、スコアが下がってしまう（過学習）

  とりうるハイパーパラメーターの値は、モデルによって異なる。

  




````
plot_model(rf, 'vc')
````

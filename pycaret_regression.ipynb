### モデルの学習状況(学習不足・過学習)を見る
#### Learning Curve(学習曲線)
x軸に学習回数、y軸に正解率をとる。

上の線が訓練データ、下がテストデータ。

正しく学習が行われていれば、学習回数が増えるに連れて、訓練データのグラフは右肩下がり、テストデータのグラフは右肩上がりに変化する。（訓練データとテストデータの正解率が近づいていく）

#### *学習回数が増えても訓練データ・テストデータ両方に対する正解率が上がらない場合*


学習不足を起こしている。説明変数の特徴量を増やすことで改善することができる。



#### *訓練データの正解率が高いまま、テストデータに対する正解率が上がらない場合*



訓練データに対する過学習を起こしている。データ数が少なく、特徴量が多い場合に過学習を起こすことが多いので、特徴量を減らすか、データ数を増やすことで解決する。




````
plot_model(rf, 'learning')
````

### Validation Curve(検証曲線)
検証曲線は、ハイパーパラメーターの値と予測性能の関係を示した折れ線グラフである。x軸は、ハイパーパラメーターの値であり、縦軸は評価指標である。評価指標は、訓練データを使った時の評価指標と検証データを使った時の評価指標の両方をプロットする。両者の指標で描かれた曲線の離れ具合で、最適なハイパーパラメーターの取りうる値を大まかに把握することができる。

例）グラフが右肩下がり
  
  パラメーターの値が大きくなるにつれて、スコアが下がってしまう（過学習）

  とりうるハイパーパラメーターの値は、モデルによって異なる。

  




````
plot_model(rf, 'vc')
````

### 残差プロット
残差の分布、または外れ値があるか確認できる。

* x軸：予測値
* y軸：残差

外れ値が発生する原因として

* 回帰式が適切ではない
* データが不適切
* 学習不足

がある


````
plot_model(rf, 'residuals')
````

###  Prediction Error Plot
予測精度を確認できる。
* x軸：実際の値
* y軸：予測値

予測が完全であれば、傾き４５°の直線に点がプロットされる。誤差が少ないほど、点は直線に近くなる。

````
plot_model(rf, 'error')
````

### Cook's Distance Plot（クックの距離）
データの外れ値を見ることができる。

クックの距離は、全てデータ用いた場合と１つのデータを除いた後求めた回帰式による予測値を用いた場合との食い違いに関する距離の測度である。

クックの距離が大きいとそのデータが回帰式による予測値に大きく影響していることを意味する。

クックの距離が0.5以上であればそのデータは異常値であると言われている。

図中の破線部分が部分が、推奨される閾値である。

````
plot_model(rf, 'cooks')
````

### Recursive Feature Selection(RFE)
特徴量の数による予測精度を確認できる。

言い換えると、予測に必要最低限な特徴量の数を確認することができると言える。

ランダムフォレストのような決定木ベースモデルでは、特徴量の数が予測制度へ与える影響は小さいが、線形モデルでは、特徴量を減らすことが重要となる。


````
plot_model(rf, 'rfe')
````
